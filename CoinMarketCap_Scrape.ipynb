```python
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver import ActionChains
import time
import csv

filename = "Coinmarket.csv"
csv_file = open(filename, 'w', newline='')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Name','Symbol','Market Cap','Price','Circulating Supply', 'Volume-24h'])


PATH = "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(PATH)
driver.get("https://coinmarketcap.com/all/views/all/")

try:
    div = WebDriverWait(driver, 5).until(
     EC.element_to_be_clickable((By.XPATH, "//*[@id='__next']/div[1]/div[2]/div/div[1]/div/div[3]/button"))
    )
    cookie = driver.find_element_by_class_name("cmc-cookie-policy-banner__close").click()
    for y in range(0,42):
        actions = ActionChains(driver)
        actions.move_to_element(div).click().perform()
        time.sleep(4)
except:
    source = driver.page_source
    content = BeautifulSoup(source,"lxml")
    table = content.findAll('table')
    for tr in table[2].findAll('tr'):
        rawname = tr.find(class_ = "cmc-table__cell cmc-table__cell--sticky cmc-table__cell--sortable cmc-table__cell--left cmc-table__cell--sort-by__name")
        if (rawname != None):
            name = rawname.text
    
        rawsymbol = tr.find(class_ = "cmc-table__cell cmc-table__cell--sortable cmc-table__cell--left cmc-table__cell--sort-by__symbol")
        if (rawsymbol != None):
            symbol = rawsymbol.text
        
        rawcap = tr.find(class_ = "cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__market-cap")
        if (rawcap != None):
            marketcap = rawcap.text
        
        rawprice = tr.find(class_ = "cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__price")
        if (rawprice != None):
            price = rawprice.text
        
        rawsupply = tr.find(class_ = "cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__circulating-supply")
        if (rawsupply != None):
            supply = rawsupply.text
    
        rawvol = tr.find(class_ = "cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__volume-24-h")
        if (rawvol != None):
            volume = rawvol.text
            csv_writer.writerow([name, symbol, marketcap, price, supply, volume])
        
csv_file.close()
        

        
        
        
```
